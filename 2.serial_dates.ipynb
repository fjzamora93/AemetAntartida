```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os as os
import math

#TODO Introducir el dataset con el que se vaya a trabajar
station = 'gdc'
df = pd.read_csv(f'./data/data_{station}/{station}_clean.csv')
df['date'] = pd.to_datetime(df['date'])

df.info()
```

### Comprobamos el número de registros por mes

```python

df_agrupacion = df.groupby(df['date'].dt.to_period('M')).count()
df_agrupacion.head(24*20)
```

```python
# Crear una secuencia de horas completa entre el mínimo y el máximo valor de 'date'
full_range = pd.date_range(start=df['date'].min(), end=df['date'].max(), freq='h')

# Reindexar el DataFrame para asegurar que todas las horas están presentes
df = df.set_index('date').reindex(full_range)
```

```python


df_agrupacion = df.groupby(df['date'].dt.to_period('M')).count()
df_agrupacion.head(24*20)
```
